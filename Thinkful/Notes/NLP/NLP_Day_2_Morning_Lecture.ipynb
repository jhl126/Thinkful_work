```
from nltk.corpus.reader.plaintext import PlaintextCorpusReader

PATH='/content/drive/My Drive/DSI Month 4/Week 13/NLP text sets/cnn_lite'
DOC_PATTERN = r'.*\.txt'
corpus = PlaintextCorpusReader(PATH, DOC_PATTERN)
```

```
# corpus.raw(NAME_OF_FILE)
# docs = [corpus.raw(fileid) for fileid in corpus.fileids()]
docs = []

for fileid in corpus.fileids():
  doc = corpus.raw(fileid)
  docs.append(doc)
```

```
from nltk.corpus import stopwords

print(stopwords.words('english'))
```

```
# no_stop = []
# for word in all_words:
#   if word.lower() not in stopwords.words('english'):
#     no_stop.append(word.lower())

no_stop = [word.lower() for word in all_words if word.lower() not in stopwords.words('english')]
```

```
len(no_stop)
```

```
no_punct = [word.lower() for word in all_words if word.isalpha()]
no_punct
```

```
no_stop_punct = [word.lower() for word in all_words if word.isalpha() if word.lower() not in stopwords.words('english')]
no_stop_punct
```

```
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer('english')
stemmed = [stemmer.stem(word) for word in no_stop_punct]
```

```
stemmed
```

```
from nltk.stem.wordnet import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in no_stop_punct]
```

```python
from nltk.corpus.reader.plaintext import PlaintextCorpusReader


PATH = '/content/drive/My Drive/DSI Month 4/Week 13/NLP text sets/oreilly_rss'
DOC_PATTERN = r'.*\.txt'
corpus = PlaintextCorpusReader(PATH, DOC_PATTERN)
```

```python
documents = [corpus.raw(fileid) for fileid in corpus.fileids()]
documents
```

```python
documents[0]
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(1,2), min_df=2)
vectors = vectorizer.fit_transform(documents)
```

```python
# get the BOW from the vectorizer
# the cow jumped over the moon
vectorizer.get_feature_names()
```

```python
count = pd.DataFrame(vectors.toarray(), 
                     columns=vectorizer.get_feature_names())
count.head()
```

```python
vectorizer = CountVectorizer(binary=True)
vectors = vectorizer.fit_transform(documents)

one_hot = pd.DataFrame(vectors.toarray(), 
                     columns=vectorizer.get_feature_names())
one_hot.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(documents)

tfidf = pd.DataFrame(vectors.toarray(), 
                     columns=vectorizer.get_feature_names())
tfidf.head()
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

tagged_documents = [TaggedDocument(doc, [i]) 
             for i, doc in enumerate(documents)]

tagged_documents
```

```python
model = Doc2Vec(tagged_documents)
```

```python
doc2vec = pd.DataFrame([[document]+list(model[document]) 
                        for document in range(len(tagged_documents))]).drop(0, axis=1)
```

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.svm import SVC

%matplotlib inline
```

```python
raw_data = pd.read_csv(
    "https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/epi_r.csv"
)
```

```python
pd.set_option("display.max_columns", None)
raw_data.head()
```

```python
raw_data.isna().mean().sort_values(ascending=False)
```

```python
# Dropping columns that are missing too many values or unnecessary
raw_data = raw_data.drop(columns=["title", "fat", "protein", "sodium", "calories"])
```

```python
raw_data.head()
```

```python
raw_data["rating"].value_counts()
```

```python
raw_data["hi_rate"] = raw_data["rating"] > 3.5
raw_data["hi_rate"] = raw_data["hi_rate"].astype(int)
raw_data.head()
```

```python
raw_data["hi_rate"].value_counts()
```

```python
X = raw_data.drop(columns=["rating", "hi_rate"])
y = raw_data["hi_rate"]
feature_names = list(X.columns.values)
```

```python
X.shape
```

```python
selector = SelectKBest(k=30).fit(X, y)
X = selector.transform(X)
scores = selector.scores_
```

```python
X.shape
```

```python
mask = selector.get_support()  # list of booleans
new_features = []  # The list of your K best features

for bool, feature in zip(mask, feature_names):
    if bool:
        new_features.append(feature)
```

```python
X = pd.DataFrame(X, columns=new_features)
```

```python
X.head()
```

```python
X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=13
)
```

```python
model = SVC(kernel="linear")
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

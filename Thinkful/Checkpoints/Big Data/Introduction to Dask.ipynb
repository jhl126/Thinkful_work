```python
from dask import delayed


@delayed
def sum_up(l):
    # takes a list of numbers as input
    return sum(l)


@delayed
def cube(n):
    return n ** 3


total = []
for i in range(1, 10):
    total.append(cube(i))

tot_sum = sum_up(total)
tot_sum
```

```python
tot_sum.visualize()
```

```python
tot_sum.compute()
```

```python
from dask import delayed

# first two fibonacci numbers
fibonacci_nums = [1, 1]

# returns the nth fibonacci number
@delayed
def get_nth_fibonacci_number(n):
    if n == 1:
        return fibonacci_nums[0]
    elif n == 2:
        return fibonacci_nums[1]
    else:
        return (
            get_nth_fibonacci_number(n - 1).compute()
            + get_nth_fibonacci_number(n - 2).compute()
        )


fib_ten = get_nth_fibonacci_number(10)
```

```python
fib_ten.visualize()
```

```python
fib_ten.compute()
```
